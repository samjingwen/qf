```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = yf.download(['AAPL', 'MSFT', 'VFINX', 'BTC-USD'], group_by='Ticker', start='2020-09-25', end='2020-10-02')
df.head()
```

```python
init_date = '2013-01-01'
symbol = 'AAPL'

df = yf.download(symbol, start=init_date, end='2020-01-01')
df = df[['Close', 'High', 'Low']].copy()

plt.figure()
df.plot()
plt.legend(['Close', 'High', 'Low'], loc='upper left') 
plt.title(symbol)
plt.ylabel('USD')
plt.xlabel('Date')
plt.grid(True)
plt.tight_layout()

plt.show()
```
